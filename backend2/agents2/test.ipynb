```python
import sys
import os

# Add backend2 directory to python path so we can import agents2
# Assuming the notebook is running from backend2/agents2/
project_root = os.path.abspath(os.path.join(os.getcwd(), '../..'))
backend2_dir = os.path.abspath(os.path.join(os.getcwd(), '..'))

if backend2_dir not in sys.path:
    sys.path.append(backend2_dir)
if project_root not in sys.path:
    sys.path.append(project_root)

try:
    from agents2.collectors.runtime_collector import RuntimeCollector
    print("Successfully imported RuntimeCollector")
except ImportError as e:
    print(f"Import failed: {e}")
    try:
        from backend2.agents2.collectors.runtime_collector import RuntimeCollector
        print("Successfully imported RuntimeCollector from backend2")
    except ImportError as e2:
        print(f"Import failed again: {e2}")

# Configuration
start_date = "2026-02-01"
end_date = "2026-02-07"

if 'RuntimeCollector' in locals():
    print(f"Running RuntimeCollector from {start_date} to {end_date}...")
    collector = RuntimeCollector(start_date=start_date, end_date=end_date)
    results = collector.run()
    
    print(f"\nCollected {len(results)} articles:")
    for item in results:
        print(f"- [{item.get('date')}] {item.get('raw_title', 'No Title')} ({item.get('source_name', 'Unknown')})")
else:
    print("RuntimeCollector class not available.")
```
